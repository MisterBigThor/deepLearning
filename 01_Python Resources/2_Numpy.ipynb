## Selección y indexado de Matrices

## Selección avanzada

```python
arr = np.arange(1,11)
arr2d = np.zeros((10,10))
arr_length = arr2d.shape[1]
for i in range(arr_length):
    arr2d[i] = i
    
arr2d
```

```python
arr2d[[2,4,6,8]]
```

```python
arr2d[[6,4,2,7]]
```

```python
arr
```

```python
arr > 4
```

```python
bool_arr = arr > 4
```

```python
arr[bool_arr]
```

```python
arr[arr > 2]
```

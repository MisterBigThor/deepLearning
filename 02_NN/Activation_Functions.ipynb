```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```
def linear_activation(x, m):
  return m*x

def ReLU(xs):
  return [0 if x < 0 else x for x in xs]

def sigmoid(xs):
  return [1 / float(1 + np.exp(-x)) for x in xs]

def tanh(xs):
  return np.tanh(xs)
```

```
x = np.linspace(-3,3,100)
y = linear_activation(x, 1)
```

```
fig, (ax, ax1, ax2, ax3) = plt.subplots(ncols=4, sharey=True, figsize=(20,10))
ax.plot(x,y, c='salmon', label='linear')
ax.set_aspect('equal')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_title('Linear activation')
ax.legend()

y = ReLU(x)
ax1.set_aspect('equal')
ax1.grid(True, which='both')
ax1.axhline(y=0, color='k')
ax1.axvline(x=0, color='k')
ax1.plot(x,y, c='salmon', label='ReLU')
ax1.set_title('ReLU activation')
ax1.legend()

y = sigmoid(x)
ax2.set_aspect('equal')
ax2.grid(True, which='both')
ax2.axhline(y=0, color='k')
ax2.axvline(x=0, color='k')
ax2.plot(x,y, c='salmon', label='Sigmoid')
ax2.set_title('Sigmoid activation')
ax2.legend()

y = tanh(x)
ax3.set_aspect('equal')
ax3.grid(True, which='both')
ax3.axhline(y=0, color='k')
ax3.axvline(x=0, color='k')
ax3.plot(x,y, c='salmon', label='TanH')
ax3.set_title('TanH activation')
ax3.legend()
```

```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm 
from sympy import symbols
from sympy import diff
import math
```

```
# 2D equation for Gradients descents Algorithms
def f(x):
	return  (x ** 4) - (4 * (x ** 2)) + 5

def df(x):
	return (4 * (x**3)) - (8 * x) 
```

```
def gradient_descent_2d(function, initial_input, learning_rate=0.02, precision= 0.0001):

	next_x = initial_input
	x_arr = [next_x]
	slope_array = [function(next_x)]

	for n in range(500):
		current_x = next_x
		gradient = function(current_x)
		next_x = current_x - learning_rate * gradient

		step_size = abs(next_x - current_x)
		x_arr.append(next_x)
		slope_array.append(function(next_x))

		if step_size < precision:
			break
	return next_x, x_arr, slope_array
```

```
x_array = np.linspace(-2, 2, 1000)


minima, x_ar , deriv_ar = gradient_descent_2d(df, 0.1)

# matplotlib graph
fig, axes = plt.subplots(figsize=(12,3))


axes.grid()
axes.plot(x_array, f(x_array), color='green', linewidth=3, alpha=0.8)
axes.scatter(x_ar, f(np.array(x_ar)), color='red', s=100, alpha=0.6) 
```

```
# 3D graph 

def t(x, y):
    r = 3 ** (-x ** 2 - y **2)
    return 1/(r + 1)
  
def dt(x, y):
    return (2*np.log(3)*x*(3 ** (-x**2 - y**2)))/((3 ** (-x**2 - y**2) + 1)**2)
```

```
def gradient_descent_3d(function, initial_input_x, initial_input_y, learning_rate=3.5, precision = 0.001):
  next_x = initial_input_x
  next_y = initial_input_y
  x_arr = [next_x]
  y_arr = [next_y]
  slope_array = [function(next_x,next_y)]

  for n in range(1500):
    current_x = next_x
    current_y = next_y
    gradient = function(current_x, current_y)
    next_x = current_x - learning_rate * gradient
    next_y = current_y - learning_rate * gradient
    
    
    x_arr.append(next_x)
    y_arr.append(next_y)
    slope_array.append(function(next_x,next_y))


  return next_x, next_y, x_arr, y_arr, slope_array
```

```
minima_x, minima_y, x_ar, y_ar, deriv_ar = gradient_descent_3d(dt, 1.5, 1.5)


x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)

x_4, y_4 =  np.meshgrid(x_4, y_4)
fig = plt.figure(figsize=(16, 12))
ax = fig.gca(projection='3d')

ax.set_xlabel('X', fontsize=20)
ax.set_ylabel('Y', fontsize=20)
ax.set_zlabel('t(x,y)', fontsize=20)

ax.plot_surface(x_4, y_4, t(x_4, y_4), cmap=cm.hot ,alpha=0.3)


ax.plot3D(x_ar, y_ar, [t(P_x, P_y) for (P_x,P_y) in zip(x_ar,y_ar)], 'gray', lw=2)

plt.show()
```

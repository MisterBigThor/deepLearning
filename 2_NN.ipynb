# Perceptron


The basic unit for deep learning. All the neurons have the same structure:
1. X0 to Xn inputs, weighted with n parameters.
2. An activation function, that calculates a value.
3. A bias value, to calibrate the model.
The inputs, plus the bias are aggregated to fill the activation function.

## The perceptron usages
The most common usages are split sets and calculate a linear regression.

## Parameter initialization
There are some init methods:
* Random init (the best)
* 0 init
* He init
* Xavier init

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

## Activation functions

```python
def identity(x):
    return x

def linear_activation(x, m=1):
    return m*x

def logistic(x):
    return (1/(1-np.exp(-x)))

def sigmoid(xs):
    return [1 / float(1 + np.exp(-x)) for x in xs]

def ReLu(xs):
    return [0 if x < 0 else x for x in xs]

def tanh(xs):
    return np.tanh(xs)
```

```python
x = np.linspace(-3,3,100)
y = linear_activation(x, 1)
fig, (ax, ax1, ax2, ax3) = plt.subplots(nrows=1,ncols=4, sharey=True, figsize=(40,10))
ax.plot(x,y, c='salmon', label='linear')
ax.set_aspect('equal')
ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_title('Linear activation')
ax.legend()

y = ReLu(x)
ax1.set_aspect('equal')
ax1.grid(True, which='both')
ax1.axhline(y=0, color='k')
ax1.axvline(x=0, color='k')
ax1.plot(x,y, c='salmon', label='ReLU')
ax1.set_title('ReLU activation')
ax1.legend()

y = sigmoid(x)
ax2.set_aspect('equal')
ax2.grid(True, which='both')
ax2.axhline(y=0, color='k')
ax2.axvline(x=0, color='k')
ax2.plot(x,y, c='salmon', label='Sigmoid')
ax2.set_title('Sigmoid activation')
ax2.legend()

y = tanh(x)
ax3.set_aspect('equal')
ax3.grid(True, which='both')
ax3.axhline(y=0, color='k')
ax3.axvline(x=0, color='k')
ax3.plot(x,y, c='salmon', label='TanH')
ax3.set_title('TanH activation')
ax3.legend()

#y = logistic(x)
#ax4.set_aspect('equal')
#ax4.grid(True, which='both')
#ax4.axhline(y=0, color='k')
#ax4.axvline(x=0, color='k')
#ax4.plot(x,y, c='salmon', label='TanH')
#ax4.set_title('Logistic activation')
#ax4.legend()
```

## Cost functions - Loss functions

### Mean Square Error (MSE/ L2 Loss)
$ MSE = \sum\limits_{i=1}^n  {(y_i - y_i^p)}^2 $ 

### Mean Absolute Error (MAE/ L1 loss)
$ MAE = \sum\limits_{i=1}^n  {|y_i - y_i^p|} $  

```python
def mse(true, pred):
    """
    true: array of true values    
    pred: array of predicted values
    
    returns: mean square error loss
    """
    
    return np.mean((true - pred)**2)

fig, ax1 = plt.subplots(1,1, figsize = (7,5))

# array of same target value 10000 times
target = 100
pred = np.arange(-10000,10000, 2)

loss_mse = [mse(target, pred[i]) for i in range(len(pred))]

# plot 
ax1.plot(pred, loss_mse)
ax1.set_xlabel('Predictions')
ax1.set_ylabel('Loss')
ax1.set_title("MSE Loss vs. Predictions")

fig.tight_layout()
```
